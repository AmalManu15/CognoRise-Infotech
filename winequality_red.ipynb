```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine_data = pd.read_csv('/content/drive/MyDrive/Cognorise/winequality-red.csv')
wine_data
```

```python
print(wine_data.head())
```

```python
print(wine_data.describe())
```

```python
print(wine_data.isnull().sum())
```

```python
plt.figure(figsize=(14, 7))
for i, column in enumerate(wine_data.columns[:-1]):
    plt.subplot(3, 4, i+1)
    sns.histplot(wine_data[column], kde=True)
    plt.title(f'{column} Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(wine_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
